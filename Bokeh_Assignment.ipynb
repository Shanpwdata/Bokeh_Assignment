Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

#Ans_2.
Glyphs are the building blocks of Bokeh visualizations. A glyph is a vectorized graphical shape or marker that is used to represent your data.

```python
#Example.
from bokeh.plotting import figure, output_file, show
x = [1,2,3,4,5]
y = [3,4,5,6,7]
output_file("Glyphs_ex.html")
p = figure(title="Bokeh Glyphs Example", x_axis_label="X-axis", y_axis_label="Y-axis")
p.circle(x, y, size=15, color="blue", alpha=0.5)
p.line(x, y, line_width=2, line_color="green", legend_label="Line")
p.square(x, y, size=12, color="red", alpha=0.7)
p.legend.title = "Glyph Type"
p.legend.label_text_font_size = "12pt"
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
#Ans_3.
from bokeh.plotting import figure, output_file, show
x = [1,2,3,4,5]
y = [3,4,5,6,7]
output_file("Glyphs_ex.html")
p = figure(title="Bokeh Glyphs Example", x_axis_label="X-axis", y_axis_label="Y-axis")
p.line(x, y, line_width=2, line_color="green", legend_label="Line")
p.legend.title = "Glyph Type"
p.legend.label_text_font_size = "12pt"
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

```python
#Ans_4.
from bokeh.plotting import curdoc, figure
from bokeh.server.server import Server
from bokeh.application import Application
from bokeh.application.handlers.function import FunctionHandler
from tornado.ioloop import IOLoop

def modify_doc(doc):
    # Create a simple scatter plot
    p = figure(title="Interactive Bokeh Server Plot", tools="pan,wheel_zoom,box_zoom,reset")

    x = [1, 2, 3, 4, 5]
    y = [6, 7, 2, 4, 5]

    p.circle(x, y, size=10)

    # Add the plot to the document
    doc.add_root(p)

# Create an application
handler = FunctionHandler(modify_doc)
app = Application(handler)

# Define the Bokeh server
server = Server({'/': app}, io_loop=IOLoop(), port=54000)
server.start()
```

```python
from bokeh.plotting import figure, show
from bokeh.embed import components

# Create a simple Bokeh plot
plot = figure(title="My Bokeh Plot", x_axis_label="X-axis", y_axis_label="Y-axis")
plot.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2)

# Generate the Bokeh script and div to embed in HTML
script, div = components(plot)
```

```python
from flask import Flask, render_template
app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run()
```
